```python
df['Genetic Risk Factor (APOE-ε4 allele)'].value_counts().plot(kind='bar', color=['blue', 'orange'], figsize=(10, 5))
plt.title('Genetic Risk Factor (APOE-ε4 allele) Distribution')
```

```python
df['Air Pollution Exposure'].value_counts().plot(kind='bar', color=['blue', 'orange'], figsize=(10, 5))
plt.title('Environmental Risk Factors Distribution')
```

```python
data = {
    'Genetic Risk Factor (APOE-ε4 allele)': df['Genetic Risk Factor (APOE-ε4 allele)'].value_counts(),
    'Air Pollution Exposure': df['Air Pollution Exposure'].value_counts()
}

# Convert the dictionary into a DataFrame
data_df = pd.DataFrame(data)

# Reset the index to make it suitable for plotting
data_df = data_df.reset_index().melt(id_vars='index', var_name='Risk Factor', value_name='Count')

# Plot the data using seaborn
sns.barplot(data=data_df, x='Risk Factor', y='Count', hue='index', palette='Set2')
plt.title('Comparison of Genetic and Environmental Risk Factors')
plt.show()
```

```python
#correlation matrix
data = {
    'Air Pollutin Exposure',
    'Genetic Risk Factor (APOE-ε4 allele)'
}
df = pd.DataFrame(data)

corr_matrix = df.corr()

print(corr_matrix)

sns.heatmap(corr_matrix, annot=True,
            cmap='coolwarm', fmt=".2f")
plt.tittle('Correlation Matrix')
plt.tight_layout()
plt.show()
```

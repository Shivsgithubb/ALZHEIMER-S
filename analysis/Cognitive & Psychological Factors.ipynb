Heatmap or Stacked bars: sleep quality correlation with diagnosis

```python
# Heatmap for Sleep Quality correlation with Alzheimer's Diagnosis
plt.figure(figsize=(8, 6))

# Convert 'Sleep Quality' to numeric values
sleep_quality_mapping = {'Poor': 1, 'Average': 2, 'Good': 3}
df['Sleep Quality Numeric'] = df['Sleep Quality'].map(sleep_quality_mapping)

# Compute the correlation matrix for relevant columns
correlation_matrix = df[['Sleep Quality Numeric', 'Alzheimer’s Diagnosis']].astype('string').corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Add labels and title
plt.title('Correlation Heatmap: Sleep Quality vs Alzheimer\'s Diagnosis', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
label_encoders = {}
for col in df_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le

# Decode labels for readability
df_plot = df.copy()
df_plot['Alzheimer’s Diagnosis'] = label_encoders['Alzheimer’s Diagnosis'].transform(df_plot['Alzheimer’s Diagnosis'])

# Convert 'Sleep Quality' to numeric values
sleep_quality_mapping = {'Poor': 1, 'Average': 2, 'Good': 3}
df['Sleep Quality Numeric'] = df['Sleep Quality'].map(sleep_quality_mapping)

# Create stacked bar
crosstab = pd.crosstab(df_plot['Sleep Quality'], df_plot['Alzheimer’s Diagnosis'])
crosstab.columns = ['No Alzheimer\'s', 'Alzheimer\'s']

# Plot
crosstab.plot(kind='bar', stacked=True, colormap='Set2')
plt.title('Alzheimer’s Diagnosis by Sleep Quality')
plt.xlabel('Sleep Quality')
#plt.xlabel('Gender')
#plt.ylabel('Number of Individuals')
plt.xticks(rotation=0)
#plt.legend(title='Diagnosis')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()
for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object':
        df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col].astype(str))

correlation_matrix = df_encoded.corr()

plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Alzheimer's Prediction Dataset")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('c:/users/Risha/Downloads/archive/alzheimers_prediction_dataset.csv')
df
```

```python
data = {
    'Income Level': df['Income Level'].value_counts(),
    'Alzheimer’s Diagnosis': df['Alzheimer’s Diagnosis'].value_counts()
}

#convert the dictionary to a DataFrame
data_df = pd.DataFrame(data)

#reset the index of the DataFrame
data_df.reset_index(inplace=True)

#plot the data using seaborn
sns.barplot(x='index', y='Income Level', data=data_df)
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.title('Income Level Distribution')
plt.show()
```

```python
# stacked bar
df.plot(kind='bar', x= 'Employment Status', y= 'Alzheimer’s Diagnosis', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Plot of Employment Status vs Alzheimer’s Diagnosis')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.show()
```

```python
#Heaat map
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

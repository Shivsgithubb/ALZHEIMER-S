```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('alzheimers_prediction_dataset.csv')
```

```python
df
```

```python
# Create histogram using Plotly
fig = px.histogram(df, 
                  x='Cognitive Test Score',
                  color='Alzheimer's Diagnosis',
                  barmode='overlay',
                  opacity=0.7,
                  color_discrete_sequence=['#2ecc71', '#e74c3c'],
                  title='Distribution of Cognitive Test Scores by Diagnosis Status',
                  labels={'Cognitive Test Score': 'Cognitive Test Score',
                         'count': 'Number of Individuals',
                         'Alzheimers Diagnosis': 'Diagnosis'})

# Update layout
fig.update_layout(
    title_x=0.5,
    plot_bgcolor='white',
    bargap=0.1,
    width=800,
    height=500
)

# Update axes
fig.update_xaxes(gridcolor='lightgray')
fig.update_yaxes(gridcolor='lightgray')

fig.show()
```
